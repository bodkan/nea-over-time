```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
ancestry = pd.read_table('../test_0.1.txt', header=None)
```

```python
plt.plot(ancestry.mean())
plt.ylim(0, 0.1)
plt.xlim(1, ancestry.shape[1])
plt.xscale('log')
```
